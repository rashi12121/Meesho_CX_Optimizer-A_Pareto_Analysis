**Task-1**

```python
import pandas as pd
import matplotlib.pyplot as plt
kyleinfo=pd.read_csv('https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/ComplaintsTable.csv')
kyleinfo
kyleinfo.head()
```

**Task-2**  [find total no. of complaints per productCategory]

```python
total_complaints=kyleinfo.groupby('ProductCategory')['ComplaintCount'].sum()
total_complaints
```

**Task-3**  [sorting the no. of complaints]

```python
sorted_complaints=total_complaints.sort_values(ascending=False)
sorted_complaints
```

**Task-4**  [finding the cummulative sum of all the Product Category]

```python
cummulative_sum=sorted_complaints.cumsum()
cummulative_sum
```

**Task-5**    [calculating the cummulative percentage of complaint each product]

```python
cummulative_percentage=round(cummulative_sum/sorted_complaints.sum()*100,2)
cummulative_percentage
```

**Task-6**  [constructing the pareto chart with bar graph]

```python
fig,ax1 = plt.subplots(figsize=(6, 3))
sorted_complaints.plot(kind='bar',color='C3',ax=ax1)
ax1.set_ylabel('Complaints')

ax2=ax1.twinx()
cummulative_percentage.plot(kind='line',color='C8',ax=ax2,marker='D')
ax2.set_ylabel('Cummulative Percentage')

plt.title('Pareto Analysis of Product Complaints')
```

**Task-7** [finding top 4 products with highest complaints]

```python
kyleinfo.head(3)
```

```python
Top4=kyleinfo.query("ProductCategory in ['Kitchenware','Bedding','Jewelry','Books']")
Top4
```

**Task-8**  [finding no of products having same complaints]

```python
sorted_returns1=Top4.groupby('ComplaintReason')['ComplaintCount'].sum().sort_values(ascending=False)
sorted_returns1
```

**Task-9**   [cummutative percentage of complaint count]

```python
cummulative_reason_percentage=round(sorted_returns1.cumsum()/sorted_returns1.sum()*100,2)
cummulative_reason_percentage
```

**Task-10**   

```python
fig,ax1 = plt.subplots(figsize=(6, 3))
sorted_returns1.plot(kind='bar', color='C3', ax=ax1)
ax1.set_ylabel('Complaints Counts')

ax2=ax1.twinx()
cummulative_reason_percentage.plot(kind='line',color='C8',ax=ax2,marker='D')
ax2.set_ylabel('Cummulative Percentage')

plt.title('Pareto Analysis of Complaints Reasons')
```
